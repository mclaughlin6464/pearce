```python
from pearce.mocks.kittens import cat_dict
```

```python
import numpy as np
```

```python
%matplotlib inline 
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
cat = cat_dict['trainingbox'](20)
```

```python
cat.load(1.0, particles=True)
```

```python
cat.populate({'logMmin':13.5}, min_ptcl=20)
```

```python
rbins = np.logspace(-.3, 1.6, 21)
rbc = (rbins[1:]+rbins[:-1])/2.0
```

```python
wp=cat.calc_wp(rbins)
```

```python
ds=cat.calc_ds(rbins)
```

```python
knncdf = cat.calc_knn_cdf(rbins, k = 1, n_ran=1e6)
```

```python
colors = sns.color_palette()
```

```python
plt.plot(rbc, wp, c=colors[0])
plt.loglog();
plt.xlabel(r'$r_p$')
plt.ylabel(r'$w_p(r_p)$')
```

```python
plt.plot(rbc, ds, c=colors[1])
plt.loglog();
plt.xlabel(r'$r_p$')
plt.ylabel(r'$\Delta \Sigma(r_p)$')
```

```python
plt.plot(rbc, knncdf, c=colors[3])
plt.loglog();
plt.xlabel(r'$r$')
plt.ylabel(r'$k$NN-CDF$(r)$')
```
